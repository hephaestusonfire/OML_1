```python
import numpy as np
import matplotlib.pyplot as plt

def linear_function(x):
    return x

def linear_derivative(x):
    return np.ones_like(x)

def nonlinear_function(x):
    return x**2

def nonlinear_derivative(x):
    return 2 * x

x = np.linspace(-10, 10, 400)

y_linear = linear_function(x)
y_linear_derivative = linear_derivative(x)
y_nonlinear = nonlinear_function(x)
y_nonlinear_derivative = nonlinear_derivative(x)

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(x, y_linear, label='Linear: f(x) = x')
plt.plot(x, y_linear_derivative, label="Linear Derivative: f'(x) = 1")
plt.xlabel('x')
plt.ylabel('f(x) and f\'(x)')
plt.title('Linear Function and Its Derivative')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(x, y_nonlinear, label='Nonlinear: f(x) = x^2')
plt.plot(x, y_nonlinear_derivative, label="Nonlinear Derivative: f'(x) = 2x")
plt.xlabel('x')
plt.ylabel('f(x) and f\'(x)')
plt.title('Nonlinear Function and Its Derivative')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
def linear_function(x):
    return x

def linear_derivative(x):
    return np.ones_like(x)

def nonlinear_function(x):
    return x**2

def nonlinear_derivative(x):
    return 2 * x

x = np.linspace(-10, 10, 400)

y_linear = linear_function(x)
y_linear_derivative = linear_derivative(x)
y_nonlinear = nonlinear_function(x)
y_nonlinear_derivative = nonlinear_derivative(x)

plt.subplot(1, 2, 2)
plt.plot(x, y_nonlinear, label='Nonlinear: f(x) = x^2')
plt.plot(x, y_nonlinear_derivative, label="Nonlinear Derivative of f'(x) + f(x)x for x=2")
plt.xlabel('x')
plt.ylabel('f(x) and f\'(x)')
plt.title('Nonlinear Function and Its Derivative')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def f_xy(x, y):
    return x**2 - y**2 + 2

def combined_expression(x, y, x0, y0):
    f_val = f_xy(x, y)
    return f_val + f_val * (x - x0) + f_val * (y - y0)

x = np.linspace(-5, 5, 400)
y = np.linspace(-5, 5, 400)
X, Y = np.meshgrid(x, y)

Z_f_xy = f_xy(X, Y)

x0, y0 = 2, 2

Z_combined = combined_expression(X, Y, x0, y0)

fig = plt.figure(figsize=(14, 6))

ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(X, Y, Z_f_xy, cmap='viridis')
ax1.set_title('Original Function: $f(x, y) = x^2 - y^2 + 2$')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('f(x, y)')

ax2 = fig.add_subplot(122, projection='3d')
ax2.plot_surface(X, Y, Z_combined, cmap='viridis')
ax2.set_title('Combined Expression: $f(x, y) + f(x, y)(x-x_0) + f(x, y)(y-y_0)$')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_zlabel('Combined Expression')

plt.tight_layout()
plt.show()
```
